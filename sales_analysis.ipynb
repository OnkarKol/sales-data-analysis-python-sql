# Visualization + Insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Explanation
* pandas â data analysis
* matplotlib â basic charts
* seaborn â better visualization

#### Load Dataset

```python
df = pd.read_csv("D:\Fusion\PROJECT\Sales_Data_Analysis_Project\cleaned_superstore.csv")
df.head()
```

#### Convert Date Column
* This converts the date column into datetime format so we can extract month/year.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

# Visualization's

## 1.Sales by Region

```python
df.groupby('Region')['Sales'].sum().plot(kind = 'bar')
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

#### Insight:
* The West region generated the highest sales among all regions.

## 2.Sales by Category

```python
df.groupby('Category')['Sales'].sum().plot(kind = 'bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

#### Insight:
* Technology category contributes the highest revenue.

## 3.Profit by Region

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.show()
```

#### Insight:
* Some regions generate high sales but relatively lower profit margins.

## 4.Monthly Sales Trend

```python
# First create month column.
df['Month'] = df['Order Date'].dt.month
```

```python
df.groupby('Month')['Sales'].sum().plot(kind = 'line',marker = 'o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

#### Insight:
* Sales fluctuate across months, with certain months showing higher demand.

## Key Business Insights
* West region generates the highest sales.

* Technology category contributes the most revenue.

* Consumer segment dominates the sales share.

* Certain months show peak sales indicating seasonal demand.
